```python
Lista = [5,3,2,1,4]

```

```python
Acum_1 = 0
Acum_2 = 0
for i in range(0,len(Lista)):
    if i%2==0:
        Acum_1 += Lista[i]
    else:
         Acum_2 += Lista[i]
print(Acum_1)
print(Acum_2)
```

```python
Acum_1-Acum_2
```

```python
Acum_1 = sum([Lista[i] for i in range(0,len(Lista)) if i%2==0])
Acum_2 = sum([Lista[i] for i in range(0,len(Lista)) if i%2!=0])
print(Acum_1-Acum_2)
```

```python
sum([Lista[i] if i%2==0 else -Lista[i] for i in range(0,len(Lista))])
```
