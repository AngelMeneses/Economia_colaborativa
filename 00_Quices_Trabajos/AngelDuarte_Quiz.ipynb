```python
## Generar el ultimo numero o elemento de la lista
def ultN(x):
    return x[-1]
ultN(Lista)
```

```python
## Generar el primer numero o elemento de la lista
def prmN(x):
    return x[0]
prmN(Lista)
```

```python
def mediana(x):
    x.sort()
    if len(x)%2!=0:
        print('Teniendo la lista: ', x, '\n', 'La mediana es: ', x[int(len(x)/2)])
    else:
        p = promedio([x[int(len(x)/2)],x[int(len(x)/2)-1]]) # Utilizando la funcion promedio anterior
        print('Teniendo la lista: ', x, '\n', 'La mediana es: ', p)
```

```python
# cuando la lista es impar
mediana(Lista)
```

```python
# par
mediana(Lista_2)
```

### Punto 4

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
dataX =  np.linspace(0,10,100)
dataY =  [ x + random.random()*2 for x in dataX ]

plt.scatter(dataX, dataY)
```

```python
result = []
for d in dataX:
    result.append([
        1,
        d,
    ])

dataX = np.array(result)

np.array(dataY)
```

```python
X = dataX
Y = dataY

Xt = np.transpose(X)

# beta OLS = (X'X)^(-1)X'Y
OLS =  np.dot(
    np.dot(
        np.linalg.inv(np.dot(Xt,X)),Xt),
        Y
)

# solución de la regresión
print('Modelo: ', OLS)
```

```python
OLS[1]
```

```python
resultY = []
for i in dataX[:,1]:
    resultY.append(OLS[0]+ OLS[1]*i)
```

```python
plt.scatter(dataX[:,1], dataY)
plt.plot(dataX[:,1], resultY)
```
