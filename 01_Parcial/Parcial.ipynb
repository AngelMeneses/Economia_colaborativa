Punto 1. Hacer una lista y encontrar la sumatoria de los números múltiplos de 3, 5 y 7

```python
Numeros = [4,1,3,8,9,3,4,12,45,12,7,86,78,15,14]

Multiplos3 = 0
Multiplos5 = 0
Multiplos7 = 0

for i in Numeros:
    if i%3==0:
        Multiplos3 = Multiplos3 +  i

    if i%5==0:
        Multiplos5 = Multiplos5 + i

    if i%7==0:
        Multiplos7 = Multiplos7 + i
        
print('La suma de los multiplos de 3 es: ', Multiplos3)
print('La suma de los multiplos de 5 es: ', Multiplos5)
print('La suma de los multiplos de 7 es: ', Multiplos7)
```

Punto 3. En el diccionario o dataframe creado anteriormente encontrar: el viaje que más distancia recorrió y el viaje que más tardó, utilizando un código.

Tiempo total del viaje

```python
for i in EcoLab.keys():
    EcoLab[i]['Tiempo Total'] = EcoLab[i]['Datos'][1] - EcoLab[i]['Datos'][0]
```

Las ubicaciones se separan en calles y carreras para generar las coordenadas para calcular distancia

```python

for i in EcoLab.keys():
    a = EcoLab[i]['Datos'][2].partition("-")[0]
    b = EcoLab[i]['Datos'][2].partition("-")[2]
    c = EcoLab[i]['Datos'][3].partition("-")[0]
    d = EcoLab[i]['Datos'][3].partition("-")[2]
    EcoLab[i]['Calles'] = [int(a[6:]),  int(c[6:])]
    EcoLab[i]['Carreras'] = [int(b[7:]),  int(d[7:])]
    
```

Las distancias totales se calculan

```python
for i in EcoLab.keys():
    EcoLab[i]['Distancia Total'] = ( (EcoLab[i]['Calles'][0]-EcoLab[i]['Calles'][1])**2 + (EcoLab[i]['Carreras'][0]-EcoLab[i]['Carreras'][1])**2)**(1/2)
```

Se calcula la mayor distancia y mayor tiempo
1. se colocan las distancias y tiempos en una lista para poder usar la función max

```python
EcoLab['Todas las Distancias'] = []
EcoLab['Todos los Tiempos'] = []
for i in Nombres:
    EcoLab['Todas las Distancias'].append(EcoLab[i]['Distancia Total'])
    EcoLab['Todos los Tiempos'].append(EcoLab[i]['Tiempo Total'])
```

```python
Dist = max(EcoLab['Todas las Distancias'])
Tiem = max(EcoLab['Todos los Tiempos'])
```

```python
for i in Nombres:
    if EcoLab[i]['Distancia Total'] == Dist:
        print('El viaje con mayor distancia es el ', i)
    if EcoLab[i]['Tiempo Total'] == Tiem:
        print('El viaje con mayor tiempo de viaje es el ', i)
```
