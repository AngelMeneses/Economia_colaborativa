## Punto 3

```python
# Se agrega al diccionario el tiempo recorrido como la resta de la hora de llegada menos la hora de salida
for i in ViajesDiarios.keys():
    ViajesDiarios[i]['Tiempo recorrido'] = ViajesDiarios[i]['Hora Llegada'] - ViajesDiarios[i]['Hora Salida']
    
```

```python
# Se agrega al diccionario las coordenadas o posiciones de cada ubicacion
## Coordenadas
Coordenadas = {
    'a': [0,0],
    'b': [1,1],
    'c': [1,2],
    'd': [2,3],
    'e': [4,4]
}
## Variables para insertar las coordenadas
Coor = ['Coor X 1', 'Coor Y 1']
Coor2 = ['Coor X 2', 'Coor Y 2']
Poss = ['a', 'b', 'c', 'd', 'e']
Lugares = ['Lugar Salida', 'Lugar Llegada']
## Se agregan al diccionario
for i in ViajesDiarios.keys():
    for k in Poss:
        if ViajesDiarios[i][Lugares[0]] == k:
            ViajesDiarios[i][Coor[0]] = Coordenadas[k][0]
            ViajesDiarios[i][Coor[1]] = Coordenadas[k][1]
    for k in Poss:
        if ViajesDiarios[i][Lugares[1]] == k:
            ViajesDiarios[i][Coor2[0]] = Coordenadas[k][0]
            ViajesDiarios[i][Coor2[1]] = Coordenadas[k][1]
```

```python
## Calculo de la distancia 
for i in ViajesDiarios.keys():
    ViajesDiarios[i]['Distancia recorrida'] = ( (ViajesDiarios[i]['Coor X 1']-ViajesDiarios[i]['Coor X 2'])**2 + (ViajesDiarios[i]['Coor Y 1']-ViajesDiarios[i]['Coor Y 2'])**2)**(1/2)
```

```python
# Con la funci√≥n Max se encuentra la mayor distancia y tiempo
Lista  = []
Lista2 = []
for i in ViajesDiarios.keys():
    Lista.append(ViajesDiarios[i]['Distancia recorrida'])
    Lista2.append(ViajesDiarios[i]['Tiempo recorrido'])
Maximo  = max(Lista)
Maximo2 = max(Lista2)
for i in ViajesDiarios.keys():
    if ViajesDiarios[i]['Distancia recorrida'] == Maximo:
        print('Viaje con mayor distancia es: ', i)
        
    if ViajesDiarios[i]['Tiempo recorrido'] == Maximo2:
        print('Viaje con mayor tiempo recorrido es: ', i)
```
